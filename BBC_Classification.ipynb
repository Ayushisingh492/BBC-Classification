```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"/content/bbc_text_cls new.csv")
```

```python
df.head()
```

```python
df['labels'].value_counts()
```

```python
df.isnull().sum()
```

```python
import re
```

```python
def remove_tag(raw_text):
    cleaned_text = re.sub(r"\n", '', raw_text)
    return cleaned_text
```

```python
df['text'] = df['text'].apply(remove_tag)
```

```python
df
```

```python
df['text'] = df['text'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
df['text'] = df['text'].apply(lambda x:[item for item in x.split() if item not in stopwords]).apply(lambda x:" ".join(x))
```

```python
df.head()
```

## tfidf


```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf= TfidfVectorizer()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

x = df.iloc[: , 0:1]
y = df['labels']

encoder = LabelEncoder()
y = encoder.fit_transform(y)

x_trai  n, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

x_train_tf = tfidf.fit_transform(x_train['text'])
x_test_tf = tfidf.transform(x_test['text'])
```

```python
x_train_tf.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
model = RandomForestClassifier()
model.fit(x_train_tf, y_train)
y_pred = model.predict(x_test_tf)
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
print(accuracy_score(y_test , y_pred))
```
